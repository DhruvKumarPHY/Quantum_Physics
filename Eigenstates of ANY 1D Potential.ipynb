```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
```

```python
N = 2000
dy = 1/N
y = np.linspace(0,1,N+1)
```

```python
# Define the Potential mL^2V
def ml2V(y):
    return 1000 * (y-1/2)**2
```

```python
V = ml2V(y)
plt.plot(y,V)
```

```python
# defining the Matrix
d = 1/dy**2 + ml2V(y)[1:-1] # diagonal with Excluding first and last terms in diagonal.
e = -1/(2*dy**2) * np.ones(len(d)-1)
```

```python
w ,v = eigh_tridiagonal(d, e)
```

```python
v
```

```python
v.T
```

```python
plt.plot(v.T[0])
plt.plot(v.T[1])
plt.plot(v.T[2])
plt.plot(v.T[3])
plt.plot(v.T[4])
# Quadratic Harmonal Potential
```

```python
# Lets take their Probality as well
plt.plot(v.T[0]**2)
plt.plot(v.T[1]**2)
plt.plot(v.T[2]**2)
plt.plot(v.T[3]**2)
plt.plot(v.T[4]**2)
```

```python
plt.bar(np.arange(0,20,1),w[0:20])
plt.ylabel('$ml^2 E/\hbar^2$')
```
